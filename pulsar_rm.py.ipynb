```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
full_pulsar_rm = pd.read_csv('~/Desktop/lv_data/pulsar_rm.csv')
full_pulsar_rm
```

```python
pulsar_rm = full_pulsar_rm[['SOURCE', 'GLON', 'GLAT', 'RM']]
pulsar_rm
```

```python
pulsar_rm = pulsar_rm[(pulsar_rm['GLON'] >= 180) | (pulsar_rm['GLON'] <= 20)]
```

```python
pulsar_rm
pulsar_rm['L>180'] = (pulsar_rm['GLON'] - 360)
```

```python
pulsar_rm
```

```python
magmo_lat = pulsar_rm[np.abs(pulsar_rm['GLAT']) <= 3.0]
```

```python
magmo_lat_pos_RM = magmo_lat[magmo_lat['RM'] > 0.]
```

```python
magmo_lat_neg_RM = magmo_lat[magmo_lat['RM'] < 0.]
magmo_lat_neg_RM
```

```python
data_P    = np.loadtxt('LB_zeeman_positive.dat')
data_N    = np.loadtxt('LB_zeeman_negative.dat')
Pos_L  = np.array(data_P[:, 0])
Pos_B  = np.array(data_P[:, 1])
Neg_L  = np.array(data_N[:, 0])
Neg_B  = np.array(data_N[:, 1])
```

```python
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(magmo_lat_pos_RM['L>180'], magmo_lat_pos_RM['GLAT'], 'bs', markersize=6, markeredgewidth=1.5, fillstyle='none', label=r'$RM$ $positive$')
ax.plot(magmo_lat_neg_RM['L>180'], magmo_lat_neg_RM['GLAT'], 'rs', markersize=6, markeredgewidth=1.5, fillstyle='none', label=r'$RM$ $negative$')
ax.plot(Pos_L, Pos_B, 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax.plot(Neg_L, Neg_B, 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
ax.axhline(y=0.0, color='black', linestyle='--')
ax.axvline(x=-25, color='k', linestyle='--')
ax.axvline(x=-35, color='k', linestyle='--')
ax.axvline(x=-44, color='k', linestyle='--')
ax.axvline(x=-56, color='k', linestyle='--')
ax.axvline(x=-66, color='k', linestyle='--')
ax.axvline(x=-76, color='k', linestyle='--')
ax.set_xlim(-180, 60.)
ax.invert_xaxis()
ax.legend(prop=dict(weight='bold', size=10.), loc='upper left')
plt.savefig('pulsar_Long_Lat.pdf', bbox_inches='tight', pad_inches=0.09, format='pdf')
```
