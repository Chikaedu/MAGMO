```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from astropy.io import fits
import astropy.units as u
from astropy.wcs import WCS
from astropy.utils.data import get_pkg_data_filename
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
```

```python
hdulist = get_pkg_data_filename('wham-ss.fits')
hdu     = fits.open(hdulist)[0]
wcs     = WCS(hdu.header)

# fig = plt.figure(figsize=(16,8))
ax = plt.subplot(projection=wcs)
ax.imshow(hdu.data, cmap='gray', vmin=1., vmax=1000., norm=LogNorm())


# ax.coords['glon'].set_ticks(color='white')
# ax.coords['glat'].set_ticks(color='white')

ax.coords['glon'].set_axislabel('Galactic Longitude')
ax.coords['glat'].set_axislabel('Galactic Latitude')
ax.set_xlim(-0.5, hdu.data.shape[1] - 0.5)
ax.set_ylim(-0.5, hdu.data.shape[0] - 0.5)
ax.grid(False)
plt.show()
# plt.savefig('h_alpha.pdf', format='pdf')
```

```python
def read_fits_file(filename):
	hdulist = fits.open(filename)
	image   = hdulist[0]
	headers = image.header
	data    = image.data

	# for header in headers:
	# 	print header, headers[header]

	naxis  = headers['NAXIS']

	nxchan = headers['NAXIS1']
	nychan = headers['NAXIS2']

	xdelt  = headers['CD1_1']
	ydelt  = headers['CD1_2']

	xrpix  = headers['CRPIX1']
	yrpix  = headers['CRPIX2']

	xrval  = headers['CRVAL1']
	yrval  = headers['CRVAL2']


	## X-axis
	xchan   = np.arange(nxchan)
	xslope  = xdelt
	xoffset = -xslope * float(xrpix)
	xaxis   = xslope*xchan + xoffset

	## Y-axis
	ychan   = np.arange(nychan)
	yslope  = ydelt
	yoffset = -yslope * float(yrpix)
	yaxis   = yslope*ychan + yoffset

	return xaxis, yaxis, data, xchan, ychan
```
