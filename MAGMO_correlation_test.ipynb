```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
magmo_df = pd.read_csv('~/Desktop/lv_data/ALL_MAGMO_fitted_params_Bfield.csv')
magmo_df
```

```python
magmo_df = magmo_df[['Source Name', 'Trans', 'LON', 'LON>180', 'LAT', 'B_field (mG)', 'Reliability']]
magmo_df
```

```python
split_pairs = magmo_df[magmo_df['Reliability'] == 'A']
```

```python
split_pairs
```

```python
split_pairs_pos = split_pairs[split_pairs['B_field (mG)'] > 0.]
split_pairs_pos
```

```python
split_pairs_neg = split_pairs[split_pairs['B_field (mG)'] < 0.]
split_pairs_neg
```

```python
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(split_pairs_pos['LON>180'], split_pairs_pos['B_field (mG)'], 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax.plot(split_pairs_neg['LON>180'], split_pairs_neg['B_field (mG)'], 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
ax.axhline(y=0.0, color='black', linestyle='--')
ax.set_xlim(-180, 60.)
ax.invert_xaxis()
ax.legend(prop=dict(weight='bold', size=10.), loc='upper left')
plt.savefig('B_field_Long.pdf', bbox_inches='tight', pad_inches=0.09, format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(split_pairs_pos['B_field (mG)'], split_pairs_pos['LAT'], 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax.plot(split_pairs_neg['B_field (mG)'], split_pairs_neg['LAT'], 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
# ax.set_xlim(-3, 3.)
ax.axhline(y=0.0, color='black', linestyle='--')
ax.axvline(x=0.0, color='black', linestyle='--')
ax.legend(prop=dict(weight='bold', size=10.), loc='upper left')
plt.savefig('B_field_scale_H.pdf', bbox_inches='tight', pad_inches=0.09, format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(np.abs(split_pairs_pos['B_field (mG)']), split_pairs_pos['LAT'], 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax.plot(np.abs(split_pairs_neg['B_field (mG)']), split_pairs_neg['LAT'], 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
# ax.set_xlim(-3, 3.)
ax.axhline(y=0.0, color='black', linestyle='--')
# ax.axvline(x=0.0, color='black', linestyle='--')
ax.legend(prop=dict(weight='bold', size=10.), loc='lower right')
plt.savefig('ABS_B_field_scale_H.pdf', bbox_inches='tight', pad_inches=0.09, format='pdf')
```

```python
fig, ax1 = plt.subplots(figsize=(14,4))
ax1.plot(split_pairs_pos['LON>180'], np.abs(split_pairs_pos['LAT']), 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax1.plot(split_pairs_neg['LON>180'], np.abs(split_pairs_neg['LAT']), 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
# ax.legend(prop=dict(weight='bold', size=10.), loc='lower right')
ax1.set_xlim(-180, 25.)
ax1.invert_xaxis()

ax2 = ax1.twinx()
ax2.plot(split_pairs_pos['LON>180'], np.abs(split_pairs_pos['B_field (mG)']), 'ro', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $positive$')
ax2.plot(split_pairs_neg['LON>180'], np.abs(split_pairs_neg['B_field (mG)']), 'b+', markersize=4, markeredgewidth=1.5, fillstyle='none', label=r'$Zeeman$ $negative$')
ax.legend(prop=dict(weight='bold', size=10.), loc='upper left')
plt.savefig('LON_B_field_scale_H.pdf', bbox_inches='tight', pad_inches=0.09, format='pdf')
```
